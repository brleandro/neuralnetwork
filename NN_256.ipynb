```python

from solution import *
import matplotlib.pyplot as plt

nn = NN(datapath='/Users/leandro/development/python/master/ml/hw3/cifar10.pkl',
            hidden_dims=(512, 256),
            lr=0.003,
            batch_size=100,
            seed=0)
plt.interactive(True)
nn.train_loop(50)


```

```python
plt.plot(list(range(50)), nn.train_logs['train_accuracy'], '-',label='Training accuracy')
plt.plot(list(range(50)), nn.train_logs['validation_accuracy'], '-',label='Validation accuracy')
plt.title(f"Accuracy evolution for 2 hidden layers NN ")
plt.legend()
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.show()
```

```python
plt.interactive(True)
plt.plot(list(range(50)), nn.train_logs['train_loss'], '-',label='Training loss')
plt.plot(list(range(50)), nn.train_logs['validation_loss'], '-',label='Validation loss')
plt.title(f"Loss evolution for 2 hidden layers NN ")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()
```
